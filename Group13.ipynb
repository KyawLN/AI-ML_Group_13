```python
# 3.0 Initial Data Preparation

# Handle Churn-related missing values
# If there is no churn category, the customer is 'Active'
# If there is no churn reason, the reason is 'Not Applicable'

df['Churn Category'] = df['Churn Category'].fillna('Active')
df['Churn Reason'] = df['Churn Reason'].fillna('Not Applicable')
```

```python
# Verify cleaning

print("Remaining Nulls:", df.isnull().sum().sum())
```

```python
# Drop unique identifiers that aren't useful for prediction

df = df.drop(columns=['Customer ID','Gender','Age','Married','Number of Dependents','City','Zip Code','Latitude','Longitude','Offer','Avg Monthly Long Distance Charges','Multiple Lines','Phone Service','Streaming TV','Streaming Movies','Streaming Music','Unlimited Data','Paperless Billing','Payment Method','Total Refunds','Total Extra Data Charges','Total Long Distance Charges'])
```

```python
# Create a binary target variable for analysis
# 1 = Churned, 0 = Stayed/Joined

df['Churn_Binary'] = df['Customer Status'].apply(lambda x: 1 if x == 'Churned' else 0)

print("Churn Distribution:")
print(df['Churn_Binary'].value_counts(normalize=True))
```

```python
# Setting 2 decimal place for Total Revenue column

df['Total Revenue'] = df['Total Revenue'].round(2)
```

```python
# Setting 2 decimal place for Total Revenue column

df['Total Charges'] = df['Total Charges'].round(2)
```

```python
# 4.0 Applying Random Forest 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, f1_score
```

```python
# Handle Churn-related missing values and create binary target

df['Churn Category'] = df['Churn Category'].fillna('Active')
df['Customer Status'] = df['Customer Status'].apply(lambda x: 'Stayed_or_Joined' if x in ['Stayed', 'Joined'] else x)
df['Churn_Binary'] = df['Customer Status'].apply(lambda x: 1 if x == 'Churned' else 0)
```

```python
# Encode the multi-class target variable

le = LabelEncoder()
df["Churn Category Encoded"] = le.fit_transform(df["Churn Category"])
```

```python
# FEATURE SELECTION & PREPROCESSING

# New, enhanced feature set based on correlation and domain knowledge

NUMERICAL_FEATURES = [
    'Tenure in Months',
    'Monthly Charge',
    'Total Charges',
    'Total Revenue',
    'Number of Referrals'
]

CATEGORICAL_FEATURES = [
    'Contract',
    'Internet Service',
    'Internet Type',
    'Online Security',
    'Premium Tech Support'
]
```

```python
# This ensures that any missing values in the chosen numerical features are removed.

df_processed = df.dropna(subset=NUMERICAL_FEATURES)
```

```python
# Define preprocessor to handle numerical and categorical features
# Numerical features are left as is for tree-based model
# Categorical features are One-Hot Encoded

preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', NUMERICAL_FEATURES),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), CATEGORICAL_FEATURES)
    ],
    remainder='drop'
)
```

```python
# Define X (features) and y (target) using the enhanced feature list

X = df_processed[NUMERICAL_FEATURES + CATEGORICAL_FEATURES]
y = df_processed["Churn Category Encoded"]
```

```python
# Train/Test Split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y)
```

```python
# 4.0 Applying Random Forest with Pipeline

# Create the Random Forest model within a pipeline

model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(
        n_estimators=200,
        max_depth=5,
        random_state=42,
        class_weight='balanced' # Added to help with target imbalance
    ))
])
```

```python
# Train the model
model_pipeline.fit(X_train, y_train)

# Predict and Evaluate
y_pred = model_pipeline.predict(X_test)
```

```python
print("--- Model Performance with Enhanced Features ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Macro Average F1-Score: {f1_score(y_test, y_pred, average='macro'):.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred, zero_division=0))
```

```python
#Reverse transform predictions back to labels

predicted_categories = le.inverse_transform(y_pred)
# Display label distribution in predictions for insight
print("\nPredicted Category Distribution (Top 5):\n", pd.Series(predicted_categories).value_counts().head())
```

```python
# Feature Importance Analysis (post-training)

print("\n--- Feature Importance ---")
```

```python
# Get feature names after One-Hot Encoding
feature_names = (
    NUMERICAL_FEATURES + 
    list(model_pipeline['preprocessor'].named_transformers_['cat'].get_feature_names_out(CATEGORICAL_FEATURES))
)

# Get feature importances from the classifier
importances = model_pipeline['classifier'].feature_importances_
feature_importances = pd.Series(importances, index=feature_names).sort_values(ascending=False).head(10)
```

```python
# Plot feature importance 

plt.figure(figsize=(12, 6))
sns.barplot(x=feature_importances.values, y=feature_importances.index, palette='viridis')
plt.title('Top 10 Feature Importance for Churn Category Prediction')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.show()

print(feature_importances)
```
